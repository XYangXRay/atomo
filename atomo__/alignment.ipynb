```python
import numpy as np
import scipy.ndimage as ndi
import tomopy
import dxchange
```

```python
data = tomopy.shepp3d(size=256)
```

```python
ang = tomopy.angles(181)
```

```python
prj = tomopy.project(data, ang)
```

```python
cc = np.zeros((181,2))
```

```python
for i in range(181):
    cc[i,0], cc[i,1] = ndi.center_of_mass(prj[i])
    
```

```python
print(np.std(cc[:,0]),np.std(cc[:,1]))
```

```python
import matplotlib.pyplot as plt
plt.scatter(cc[:,0], cc[:,1])
```
